# Matplotlib
-  다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

### 1. 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
```

```python
#  Title설정 
plt.plot(x,y)
plt.title("Line Graph")
```

```python
#  Title설정 - 한글  
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'        # 다른 글씨체 설정 가능 
matplotlib.rcParams['font.size'] = 15                       # 글자 크기 지정
matplotlib.rcParams['axes.unicode_minus'] = False           # 한글폰트 사용시, 마이너스 글자 깨지는 현상 해결
plt.plot(x,y)
plt.title("선 그래프")
```

```python
# 마이너스 축 표시 
plt.plot([-1,0,1],[-5,-1,2])
```
