```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'       
matplotlib.rcParams['font.size'] = 15                     
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2] 
labels = ["python","java","js","c#","c/c++",'etc']

# values : 값 , 
# labels : 각 데이터 라벨, 
# autopct : 비율 , 
# startangle : 각도 조절, 
# counterclock : 시계반대방향 여부
plt.pie(values, labels=labels, autopct="%.1f", startangle=90, counterclock=True)              
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2] 
labels = ["python","java","js","c#","c/c++",'etc']
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.05] * 6     # 균등하게 띄우기

# explode : 각 영역을 띄우기
plt.pie(values, labels=labels, explode=explode)              
plt.show()
```

```python
plt.pie(values,labels=labels,explode=explode)
plt.legend(loc=(1.2, 0.3))      # 범례 위치 조정
plt.show()
```

```python
plt.pie(values,labels=labels,explode=explode)
plt.legend(loc=(1.2, 0.3), title="언어별 선호도")      # title : 범례 제목 설정
plt.title("언어")
plt.show()
```
