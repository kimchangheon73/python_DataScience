# 3.  Index
- 데이터에 접근할 수 있는 주소 값

```python
import pandas as pd
data = {
    "이름" : ["탄지로","이노스케","젠이츠","네즈코","무잔","코쿠시보","도우마","아카자"],
    "소속" : ["귀살대","귀살대","귀살대","귀살대","오니","오니","오니","오니"],
    "기술" : ["해의호흡","짐승의호흡","번개의호흡","폭혈","촉수공격","달의호흡","빙결","무술"],
     '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df=pd.DataFrame(data, index=["1번","2번","3번","4번","5번","6번","7번","8번"])
df
```

```python
# 인덱스 확인
df.index
```

# index 이름 설정

```python
df.index.name = "지원번호"
df
```

### index 초기화

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # 원래 스던 인덱스 삭제
```

```python
df.reset_index(drop=True,inplace=True) # 실제 데이터에 바로 반영
```

```python
df
```

### index 설정 
- 지정한 컬럼으로 인덱스 설정

```python
df.set_index("이름")
```

```python
df.set_index("이름",inplace=True)
df
```

### index 정렬
- 내림차순/오름차순으로 정렬

```python
df.sort_index()   # 인덱스로 오름차순 정렬
```

```python
df.sort_index(ascending=False)    # 내림차순 정렬
```
