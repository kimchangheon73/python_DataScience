# 2. DataFrame
- 2차원 데이터 (Series들의 모임)

### Data준비
- 사전 (dict) 자료구조를 통해 생성 
- 예시 : 귀멸의 칼날 인물 8명에 대한 데이터

```python
data = {
    "이름" : ["탄지로","이노스케","젠이츠","네즈코","무잔","코쿠시보","도우마","아카자"],
    "소속" : ["귀살대","귀살대","귀살대","귀살대","오니","오니","오니","오니"],
    "기술" : ["해의호흡","짐승의호흡","번개의호흡","폭혈","촉수공격","달의호흡","빙결","무술"],
     '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
# 데이터 접근  df['칼럼 명']
df['이름']
```

```python
# 2개 이상의 칼럼 데이터 접근 -- > 리스트로 감싼 형태로 전달 
df[["이름","기술"]]
```

### DataFrame 객체 생성 (index 지정)
- 인덱스 갯수가 안맞으면 에러가 발생함

```python
df=pd.DataFrame(data, index=["1번","2번","3번","4번","5번","6번","7번","8번"])
df
```

### DataFrame 객체 생성 ( columns 지정)
- 원하는 데이터 컬럼만 선택하거나 순서를 변경

```python
df = pd.DataFrame(data, columns=["이름","키","기술"])
df
```
