# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'       
matplotlib.rcParams['font.size'] = 15                     
matplotlib.rcParams['axes.unicode_minus'] = False
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)    
```

```python
# 선 두깨 설정
plt.plot(x,y, linewidth= 5)
```

# Marker

```python
# 마커 변경 -- > o , ^ , x , 
plt.plot(x,y,marker="o")
```

```python
# 선 스타일 지정
plt.plot(x,y, marker="o",linestyle="None")
```

```python
# 마커 스타일 지정 - 모양, 크기, 테두리, 색상
plt.plot(x,y, marker="o",markersize=20, markeredgecolor="red", markerfacecolor="yellow")
```

# 선 스타일

```python
# 세밀한 점섬
plt.plot(x,y, linestyle=":")
```

```python
plt.plot(x,y, linestyle="--")
```

```python
plt.plot(x,y, linestyle="-.")
```

```python
# 선 색상 변경
plt.plot(x,y,color="pink")
```

## 포멧

```python
plt.plot(x,y,'ro--')            # color. marker, linestyle을 한번에 지정
```

```python
plt.plot(x,y,'b^:')         # 파란선, 삼각형 마커, 점선표시
```

```python
plt.plot(x,y,'go')          # 초록색선, 동그라미 마커, 선표시 x
```

## 축약어

```python
plt.plot(x,y,marker="o",mfc="red",ms=10,mec="blue",ls=":")
```

## 투명도

```python
plt.plot(x,y,marker="o",mfc="red",ms=10,mec="blue",ls=":", alpha=0.3)
```

# 그래프 크기

```python
plt.figure(figsize=(10,5))      # figsize =(가로, 세로)
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=100)      # dpi = 인치당 도트수 (해상도) - 확대 하는 기능
plt.plot(x,y)
```

# 배경색

```python
plt.figure(facecolor="yellow")
plt.plot(x,y)
```
