# 4. 파일 저장 및 열기 
- DataFrame을 Excel, csv, txt 형태의 파일로 저장 및 열기

```python
import pandas as pd
data = {
    "이름" : ["탄지로","이노스케","젠이츠","네즈코","무잔","코쿠시보","도우마","아카자"],
    "소속" : ["귀살대","귀살대","귀살대","귀살대","오니","오니","오니","오니"],
    "기술" : ["해의호흡","짐승의호흡","번개의호흡","폭혈","촉수공격","달의호흡","빙결","무술"],
     '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df=pd.DataFrame(data, index=["1번","2번","3번","4번","5번","6번","7번","8번"])
df.index.name="지원번호"
df
```

```python
# 저장하기 - csv파일로 ( 한글 : encoding="utf-8-sig" ) 
df.to_csv("score.csv",encoding="utf-8-sig", index=False)
```

```python
 # 저장하기 -txt파일로 ( 탭으로 구분된 텍스트 파일 )
 df.to_csv("score.txt",sep="\t")
```

```python
# 저장하기 - excel파일로 
df.to_excel("score.xlsx")
```

```python
 # 열기 - csv파일 열기 
 pd.read_csv("score.csv")
```

```python
# 지정된 갯수만큼의 row를 건너띔
pd.read_csv("score.csv",skiprows=1)
```

```python
# 지정된 1,3,5 row를 건너띔
pd.read_csv("score.csv",skiprows=[1,3,5])
```

```python
# 지정된 갯수만큼의 row만 가져옴 
pd.read_csv("score.csv",nrows=4)
```

```python
# 처음 2개 무시하고 4개를 가져옴 - nrows는 데이터의 맨 위에 있는 row를 칼럼으로 사용한다 
pd.read_csv("score.csv",skiprows=2, nrows=4)
```

```python
# 텍스트 파일 열기 - 구분자 지정
pd.read_csv("score.txt",sep="\t")
```

```python
# 텍스트 파일 열기 - 구분자 지정, 인덱스 칼럼 지정 
pd.read_csv("score.txt",sep="\t", index_col="지원번호")
```

```python
# 엑셀 파일 열기 
pd.read_excel("score.xlsx")
```

```python
pd.read_excel("score.xlsx",index_col="지원번호")
```
