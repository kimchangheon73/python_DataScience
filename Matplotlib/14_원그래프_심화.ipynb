# 14. 원 그래프 심화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'       
matplotlib.rcParams['font.size'] = 15                     
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2] 
labels = ["python","java","js","c#","c/c++",'etc']
colors = ['r','g','b','c','m','y']
explode = [0.05]*6
plt.pie(values, 
        labels=labels,              # 범례 
        autopct="%.1f",             # 퍼센트 표시
        startangle=90,              # 시작 각도
        counterclock=False,         # 시계반대방향 여부
        colors=colors,              # 색상변경
        explode=explode)            # 각 영역 띄어쓰기
plt.show()
```

```python
wedgeprops = {"width" : 0.2}
plt.pie(values, 
        labels=labels,              # 범례 
        autopct="%.1f",             # 퍼센트 표시
        startangle=90,              # 시작 각도
        counterclock=False,         # 시계반대방향 여부
        colors=colors,              # 색상변경
        explode=explode,            # 각 영역 띄어쓰기
        wedgeprops=wedgeprops       # 도넛 모양의 원 그래프 표시 ( 먼치킨 크기를 지정 )  
)
plt.show()
```

```python
wedgeprops = {"width" : 0.8}
plt.pie(values, 
        labels=labels,              # 범례 
        autopct="%.1f",             # 퍼센트 표시
        startangle=90,              # 시작 각도
        counterclock=False,         # 시계반대방향 여부
        colors=colors,              # 색상변경
        explode=explode,            # 각 영역 띄어쓰기
        wedgeprops=wedgeprops       # 속성을 정의하는 부분  
)
plt.show()
```

```python
wedgeprops = {"width" : 0.8, "edgecolor":"w", "lw":"3"}
plt.pie(values, 
        labels=labels,              # 범례 
        autopct="%.1f%%",           # 퍼센트 표시
        startangle=90,              # 시작 각도
        counterclock=False,         # 시계반대방향 여부
        colors=colors,              # 색상변경
        explode=explode,            # 각 영역 띄어쓰기
        wedgeprops=wedgeprops       # width : 도넛모양정도,     edgecolor : 테두리 색,      lw : 테두리 선 두께  
)
plt.show()
```

# 10% 이상인 데이터만 글자가 보이게끔 한다

```python
# 10% 이상인 데이터만 글자가 보이게끔 한다
def custom_autopct(pct):
    return ("%.0f%%"%pct) if pct>=10 else ""

wedgeprops = {"width" : 0.6, "edgecolor":"w", "lw":"3"}
plt.pie(values, 
        labels=labels,              
        autopct=custom_autopct,           
        startangle=90,              
        counterclock=False,         
        colors=colors,              
        explode=explode,            
        wedgeprops=wedgeprops,
        pctdistance=0.7     # 중간으로부터 퍼센트표시가 멀리 떨어져서 표시되는 정도(0.0 ~ 1.0)
)
plt.show()
```

## DataFrame  활용

```python
import pandas as pd
df = pd.read_excel("../Pandas/score.xlsx")
df
```

```python
grp = df.groupby("소속").size()
wedgeprops = {"width" : 0.6, "edgecolor":"w", "lw":"3"}

plt.pie(list(grp.values), 
        labels=list(grp.index),
        autopct="%.0f%%",
        startangle=90,
        counterclock=False,
        explode=[0.05]*2,
        wedgeprops=wedgeprops)
plt.title("소속")
plt.legend(list(grp.index), loc = (1.2,0.5), title="소속")
```
