# 12. 데이터 수정

```python
import pandas as pd

df = pd.read_csv("./score.csv",index_col="지원번호")
df
```

### 컬럼 수정

```python
df['소속'].replace({"오니":"상현"},inplace=True)
```

```python
df.SW특기 = df.SW특기.str.lower()                       # sw 문자열을 모두 소문자로 수정
df
```

```python
df.소속 = df.소속+" 소속"                                # 소속 데이터 마다 '소속' 붙히기
df
```

### Columns 추가


```python
df['총합'] = df.loc[:,"국어":"사회"].sum(axis=1)
df
```

### 조건에 따른 데이터 추가/수정

```python
# 결과 컬럼에 총합이 400이상이면 '우수' 이하면 '미흡'데이터를 추가
df['결과'] =  "미흡"
df.loc[ df['총합'] > 400, '결과' ] = "우수"
df
```

### Columns  삭제

```python
df.drop(columns=["총합"])
```

```python
df.drop(columns=["국어","영어","수학"])
```

### Row 삭제 

```python
df.drop(index="4번")
```

```python
# 수학점수가 80점 미만인 학생 데이터 삭제
con1 = df['수학']<80
print(df[con1].index)
df.drop(index=df[con1].index)
```

### Row 추가 

```python
df.loc["9번"] = ["렌고쿠 쿄주로","귀살대 소속","화염의 호흡",200,90,90,90,90,90,"Kotlin",450,"우수"]
df
```

### cell 수정

```python
# 데이터 한개 수정
df.loc['4번',"SW특기"] ="Python"
df
```

```python
# 데이터 여러개 수정
df.loc["5번",["소속","SW특기"]] = ["오니 대장","C++"]
df
```
