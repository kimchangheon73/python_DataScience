# 2. 출생아 수 및 합계춢산율

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'       
matplotlib.rcParams['font.size'] = 15                     
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel("./stat_101101.xls",header=2, usecols="B:K",index_col="Unnamed: 1").iloc[[0,2]]
df
```

```python
df.index            # 인덱스 확인 
df.index.values     # 정확한 인덱스 value 확인
```

```python
df.rename(index={"출생아수(명)" : "출생아 수", "합계출산율(가임여성1명당\xa0명)" : "합계 출산율"}, inplace=True)
df
```

```python
df = df.T
df
```

```python
df['출생아 수'] = df['출생아 수'].str.replace(",",".").astype("float64").round(1)
df['출생아 수']
```

```python
df['합계 출산율'] = df['합계 출산율'].astype("float64")
```

```python
# 서로 다른 범위를 가지기 때문에 데이터 시각화 효율이 좋지 않음
plt.plot(list(df.index), df['출생아 수'])
plt.plot(list(df.index), df['합계 출산율'])
```

```python
fig, axs = plt.subplots(figsize=(15,10))
axs.plot(df.index, df['출생아 수'], color="#ff812d")

ax2 = axs.twinx()       # x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color = "#ffd100")
```

```python
fig, axs = plt.subplots(figsize=(15,5))
axs.set_ylabel("출생아 수(천명)")
axs.set_ylim(250,700)
axs.set_yticks([300,400,500,600])
axs.bar(df.index, df['출생아 수'], color="#ff812d")

ax2 = axs.twinx()
ax2.set_ylabel("합계 출산율(가임여성 1명당 명")
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color = "#ffd100", lw=5)
```

```python
fig, axs = plt.subplots(figsize=(15,5))
axs.set_ylabel("출생아 수(천명)")
axs.set_ylim(250,700)
axs.set_yticks([300,400,500,600])
axs.bar(df.index, df['출생아 수'], color="#ff812d")
for idx, val in enumerate(df['출생아 수']):
    axs.text(idx, val+12, val, ha="center")

ax2 = axs.twinx()
ax2.set_ylabel("합계 출산율(가임여성 1명당 명")
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color = "#ffd100", lw=5)
```
