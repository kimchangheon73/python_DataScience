# 14. 그룹화

```python
import pandas as pd

df = pd.read_csv("./score.csv",index_col="지원번호")
df
```

```python
df.groupby("소속")
```

```python
df.groupby("소속").get_group("귀살대")
```

```python
df.groupby("소속").get_group("오니")
```

```python
df.groupby("소속").mean()               # 계산가능한 데이터들의 평균값
```

```python
df.groupby("소속").size()               # 각 그룹의 크기
```

```python
df.groupby("소속").size()['귀살대']     # 소속으로 그룹화 한 뒤 귀살대에 해당하는 데이터의 수 
```

```python
df.groupby("소속")['키'].mean()         # 그룹화 한뒤 키에 대한 평균 데이터
```

```python
df.groupby("소속")['수학',"영어","국어"].mean()
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
# 소속과 학년으로 그룹화
df.groupby(["소속","학년"]).mean()
```

```python
df.groupby("학년").mean()       # 학년별 평균 데이터
```

```python
df.groupby("학년").mean().sort_values("키",ascending=False)     # 그룹화 데이터의 정렬
```

```python
df.groupby(['소속','학년']).sum()
```

```python
# 학년별 sw특기생의 수
df.groupby(["소속"])['SW특기'].count()
```

```python
party = df.groupby("소속")
party['학년'].value_counts()            # 소속으로 그룹화를 한 뒤에 학년별 학생 수를 가져옴?
```

```python
party['학년'].value_counts().loc["귀살대"]          # 소속으로 그룹화를 한 뒤에 귀살대에 대해서 학년별 학생 수를 가져옴
```

```python
party['학년'].value_counts().loc["오니"]            # 소속으로 그룹화를 한 뒤에 오니에 대해서 학년별 학생 수를 가져옴
```

```python
party['학년'].value_counts(normalize=True).loc['귀살대']         # 소속으로 그룹화를 한 뒤에 귀살대에 대한 학년 별 전체 %를 구함 
```

```python
party['학년'].value_counts(normalize=True).loc['오니']         # 소속으로 그룹화를 한 뒤에 오니에 대한 학년 별 전체 %를 구함 
```
