# Covid - 19 백신 종류별 접종 인구

```python
days = [1, 2, 3]        # 일자
az = [2, 4, 8]          # 단위 : 만명 (1일부터 3일가지 아스트라제네카 접종인구)
pfizer = [5,1,3]        # 화이자
moderna = [1,2,5]       # 모더나

plt.plot(days, az, marker="d")
plt.plot(days, pfizer, marker="o",ls="--")
plt.plot(days, moderna,marker="^",ls=":")
plt.legend(['az','pfizer','moderna'], ncol=3)       # 개의 컬럼만큼 범례를 표시 (옆으로)
```
