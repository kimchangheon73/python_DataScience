```python
import pandas as pd

df = pd.read_csv("./score.csv",index_col="지원번호")
df
```

# 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm를 붙이는 함수
def add_cm(height):
    return str(height)+'cm'


# 키 데이터에 대해서 위에 함수를 호출한 결과 데이터를 반영함
df.키 = df.키.apply(add_cm)
df
```

```python
# Sw특기에서 첫글자는 대문자 나머지는 소문자로 데이터를 수정
# capitalize : str함수 중 첫글자는 대문자로 나머지는 소문자로 만드는 매서드
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()        #첫 글자는 대문자로, 나머지는 소문자로
    else:
        return lang
    
df.SW특기 = df.SW특기.apply(capitalize)
df
```

```python
# 소속이 '귀살대'라면 주인공, 아니면 악역으로 '역할'을 표시하라
def position(types):
    if types =="귀살대":
        return "주인공"
    else:
        return "악역"
    
df['역할'] = df['소속'].apply(position)
df
```
