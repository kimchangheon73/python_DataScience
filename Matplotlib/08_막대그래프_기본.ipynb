# 8. 막대그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'       
matplotlib.rcParams['font.size'] = 15                     
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ["탄지로","젠이츠","이노스케"]
values= [190,187,184]
plt.bar(labels,values)
```

```python
labels = ["탄지로","젠이츠","이노스케"]
values= [190,187,184]
plt.bar(labels,values,color="red")
```

```python
labels = ["탄지로","젠이츠","이노스케"]
values= [190,187,184]
color=['r','g','b']
plt.bar(labels,values,color=color)          # 값마다 여러 색 지정하기
```

```python
labels = ["탄지로","젠이츠","이노스케"]
values= [190,187,184]
color=['r','g','b']
plt.bar(labels,values,color=color, alpha=0.5)   # 투명도 조절
```

```python
labels = ["탄지로","젠이츠","이노스케"]
values= [190,187,184]
color=['r','g','b']
plt.bar(labels,values,color=color)
plt.ylim(150,200)                       # 키가 150부터 200인 사이의 정보만 표시 
```

```python
plt.bar(labels,values, width=0.5)       # bar 두께 조절
```

```python
plt.bar(labels,values,width=0.3)
plt.xticks(rotation=45)             # x축의 데이터 각도를 조절
plt.yticks(rotation=45)             # y축의 데이터 각도를 조절
```

```python
labels = ["탄지로","젠이츠","이노스케"]
values= [190,187,184]
ticks = ['1번학생','2번학생','3번학생']
plt.bar(labels,values)
plt.xticks(labels,ticks)                # labels 데이터를 우리가 지정한 ticks데이터로 변경하고 싶을 때 지정 
```
