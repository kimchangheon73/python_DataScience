# 8. 데이터 선택 (iloc)
- 위치를 이용해 원하는 row에서 원하는 columns를 선택

```python
import pandas as pd

df = pd.read_csv("./score.csv",index_col="지원번호")
df
```

```python
df.iloc[0:5]            # 0~4번째 위치의 데이터
```

```python
df.iloc[0,1]            # 0번째 위치의 데이터 중 첫번재 인덱스 컬럼의 데이터
```

```python
df.iloc[4,2]            # 4번째 위치의 데이터 중 2번째 인덱스 컬럼의 데이터 
```

```python
df.iloc[[0,1],2]        # 0,1번째 위치의 데이터 중 2번째 인덱스 컬럼 데이터
```

```python
df.iloc[[0,1],[3,4]]        # 0,1번째 위치의 데이터 중 3,4번째 인덱스 컬럼 데이터
```

```python
df.iloc[0:5,3:8]            # 0~4번째 위치의 학생 중 인덱스 칼럼이 3~8인 데이터
```
